```python
data = df.drop(['date','lights'],axis=1)
```

```python
data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normalizer = MinMaxScaler()
```

```python
new_data = pd.DataFrame(normalizer.fit_transform(data), columns=data.columns)
```

```python
new_data.head(5)
```

```python
x_features = new_data.drop(columns=['Appliances'])
y_target = new_data['Appliances']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_features, y_target,
test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_prediction = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = np.round(mean_absolute_error(y_test, y_prediction),2)
```

```python
print('Mean Absolute Error: ', mae)
```

```python
residual_ss = np.round(np.sum(np.square(y_test - y_prediction)),2)
```

```python
print('RSS: ', residual_ss)
```

```python
rmse = np.round(np.sqrt(mean_squared_error(y_test, y_prediction)),3)
```

```python
print('Root Mean Square Error: ', rmse)
```

```python
r2 =  np.round(r2_score(y_test, y_prediction),2)
```

```python
print('Coeeficient: ',r2)
```

```python
model_feature = model.coef_
```

```python
for i, (name,coef) in enumerate(zip(x_features.columns, model_feature)):
    print(f'Column Name {i+1}: {name}, Coefficient: {coef}')
    print('\n')
```
